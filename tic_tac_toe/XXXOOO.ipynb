# Игра "крестики-нолики"
За основу взял код из вебинара *Разбор итогового практического задания 6.1*,<br>
но решил сделать так, что бы человек играл не сам с собой,    
а играл против компьютера.    
Поэтому пришлось написать функцию <code>ai()</code>, отвечающую за ход компьютера    
и  внести изменения в игровой цикл.

```python
# Создаём поле
field = [[" "] * 3 for i in range(3) ]

# Приветствие
def greet():
    '''Приветствие и правила игры'''
    print("-------------------")
    print("  Приветсвуем вас  ")
    print("      в игре       ")
    print("  крестики-нолики  ")
    print("-------------------")
    print(" формат ввода: x y ")
    print(" x - номер строки  ")
    print(" y - номер столбца ")

# Выводим поле на экран
def show():
    '''Выводим изображение игрового поля в консоль'''
    print()
    print("    | 0 | 1 | 2 | ")
    print("  --------------- ")
    for i, row in enumerate(field):
        row_str = f"  {i} | {' | '.join(row)} | "
        print(row_str)
        print("  --------------- ")
    print()
    
# Проверка выигрышных комбинаций
def check_win():
    '''Проверка, что на поле есть выигрышная комбинация.
    На выходе: True - если есть, False - если нет'''
    win_cord = (((0, 0), (0, 1), (0, 2)), 
                ((1, 0), (1, 1), (1, 2)), 
                ((2, 0), (2, 1), (2, 2)),
                ((0, 2), (1, 1), (2, 0)), 
                ((0, 0), (1, 1), (2, 2)), 
                ((0, 0), (1, 0), (2, 0)),
                ((0, 1), (1, 1), (2, 1)), 
                ((0, 2), (1, 2), (2, 2)))
    for cord in win_cord:
        symbols = []
        for c in cord:
            symbols.append(field[c[0]][c[1]])
        if symbols == ["X", "X", "X"]:
            return True
        if symbols == ["O", "O", "O"]:
            return True
    return False

# Спрашиваем координаты у пользователя
def ask():
    '''Спрашиваем координаты у пользователя и проводим проверку,
    что введено 2 значения, они являются числами,
    находятся в заданном диапазоне и клетка свободна'''
    while True:
        cords = input("  Ваш ход ( строка столбец ) : ").split()
        
        if len(cords) != 2:
            print(" Введите 2 координаты! ")
            continue
        
        x, y = cords
        
        if not(x.isdigit()) or not(y.isdigit()):
            print(" Введите числа! ")
            continue
        
        x, y = int(x), int(y)
        
        if 0 > x or x > 2 or  0 > y or  y > 2 :
            print(" Координаты вне диапазона! ")
            continue
        
        if field[x][y] != " ":
            print(" Клетка занята! ")
            continue
        
        return x, y   
```

### Алгоритм игрового интеллекта
Первым делом **ИИ** пробегает по всему полю и если найдёт пустую клетку,      
временно ставит там **"O"** и с помощью функции <code>check_win()</code>   
проверяет, приносит ли это победу.   
Если этот ход приносит победу, то **ИИ** возвращает победные координаты <code>return x,y</code>    

Если победный ход не обнаружен,     
то ИИ пробегает по всему полю и если найдёт пустую клетку,      
временно ставит там **"X"** и с помощью функции <code>check_win()</code>   
проверяет, приносит ли это победу **врагу**, то есть Вам.    
Если этот ход приносит вам победу, то ИИ возвращает эти координаты <code>return x,y</code>,<br>
что бы поставить туда **"O"** и не дать вам победить.   

Если и такой ход не обнаружен,     
то разочарованный **ИИ** возвращает координаты первой попавшейся пустой клетки

```python
# Игровой интеллект, играющий за нулики
def ai():
    # 1 цикл. Ищем победный ход
    for x in range(0,3):
        for y in range(0,3):            
            if field[x][y] == " ":
                field[x][y] = "O"     # временно ставим "O"
                if check_win(): 
                    field[x][y] = " " # подчищаем после экспериментов
                    return x,y        # нашли победный ход
                field[x][y] = " "     # подчищаем после экспериментов
                
    # 2 цикл. Ищем победный ход врага            
    for x in range(0,3):
        for y in range(0,3):
            if field[x][y] == " ":
                field[x][y] = "X"     # временно ставим "X"
                if check_win():
                    field[x][y] = " " # подчищаем после экспериментов
                    return x,y        # не даём врагу победить
                field[x][y] = " "     # подчищаем после экспериментов
                
    # 3 цикл. Если ничего не нашли,            
    for x in range(0,3):
        for y in range(0,3):
            if field[x][y] == " ":    # возвращаем первую найденную пустую клетку 
                return x,y
```

## Игровой цикл

```python
greet()
field = [[" "] * 3 for i in range(3) ]
count = 0
while True:
    count += 1
    show()
    
    # ваш ход ----------------------
    x, y = ask()
    field[x][y] = "X"    
    if check_win():
        show()
        print( "Ура!!! Вы победили!" )
        break
    if count == 5:
        show()
        print( "Ничья!" )
        break    
    
    # ход компьютера ----------------
    x, y = ai()
    field[x][y] = "O"
    if check_win():
        show()
        print( "Это позор :(" )
        break
```
